```python
test_results = ["pass", "pass", "fail", "pass", "fail", "pass"]
total = len(test_results)
passed = test_results.count("pass")
failed = test_results.count("fail")

print(f"AI Test Results Summary")
print(f"Total tests: {total}")
print(f"Passed: {passed} ({round(passed/total*100)}%)")
print(f"Failed: {failed} ({round(failed/total*100)}%)")
```
