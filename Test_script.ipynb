```python
# =============================================
# Day 9 Practice — AI Test Evaluator
# =============================================

# Test cases — prompt, expected output, actual output
test_cases = [
    {"id": 1, "expected": "Paris",       "actual": "Paris"},
    {"id": 2, "expected": "Shakespeare", "actual": "Shakespeare"},
    {"id": 3, "expected": "8",           "actual": "9"},
    {"id": 4, "expected": "Water",       "actual": "H2O"},
    {"id": 5, "expected": "Sun",         "actual": "Sun"},
    {"id": 6, "expected": "Dog",         "actual": "dog"}
]

passed = 0
failed = 0
failed_cases = []

# Loop through every test case
for test in test_cases:
    if test["expected"].lower() == test["actual"].lower():
        print(f"Test {test['id']}: PASS ✓")
        passed += 1
    else:
        print(f"Test {test['id']}: FAIL ✗ — Expected '{test['expected']}' got '{test['actual']}'")
        failed += 1
        failed_cases.append(test["id"])

# Summary
total = passed + failed
pass_rate = round((passed / total) * 100)

print("\n" + "=" * 45)
print("EVALUATION SUMMARY")
print("=" * 45)
print(f"Total: {total} | Passed: {passed} | Failed: {failed}")
print(f"Pass Rate: {pass_rate}%")

if pass_rate >= 90:
    print("Status: EXCELLENT")
elif pass_rate >= 70:
    print("Status: ACCEPTABLE — monitor closely")
else:
    print("Status: FAIL — review required")

if failed_cases:
    print(f"Failed Test IDs: {failed_cases}")

```

```python
# Writing to a .txt file
with open("test_results.txt", "w") as file:
    file.write("AI Test Results\n")
    file.write("===============\n")
    file.write("   Test 1: PASS\n\n")
    file.write("Test 2: FAIL   \n ")

print("File saved!")

# Reading a .txt file
with open("test_results.txt", "r") as file:
    content = file.read()
    print(content)

#Reading line by line
with open("test_results.txt","r") as file:
    for line in file:
        print(line.strip())
```
